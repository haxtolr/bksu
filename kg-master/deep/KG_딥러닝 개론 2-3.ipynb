```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/test4 (1).csv')

print(df.columns)
print(df.head(3))
print(df.shape)
```

```python
#판다스 -> 넘파이 2차원 배열로 재구성
df_x = np.array(df['x']).reshape(-1, 1)
df_y = np.array(df['y']).reshape(-1, 1)
df_z = np.array(df['z']).reshape(-1, 1)

print(df_x[:3])
print(df_x)
```

```python
x = np.arange(0, 10)
plt.scatter(df_x, df_y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
data = np.column_stack((df_x, df_y))
```

# 의사 결정나무 모델 생성후 학습
- Sklearn으로 데이터 전처리 및 학습



```python
from sklearn import tree
```

```python
from sklearn.model_selection import train_test_split

train_input , test_input , train_output , test_output = train_test_split(data, df_z)

# train_input, test_input, train_output = train_test_split(data, df_z, test_size= 0.2)
# train_input, test_input, train_output = train_test_split(data, df_z, train_size=0.8)
```

```python

from sklearn.tree import DecisionTreeClassifier, plot_tree

DT = DecisionTreeClassifier(max_depth=2) # 아래로 2층만 생성
```

```python
# 모델 학습
DT.fit(train_input, train_output)
```

```python
# 모델 검증
DT.score(test_input, test_output) # 1.0이면 100점
```

```python
# 트리 모델 시각화
# (시각화할 모델명, 트리 최대 깊이, 색채우기, 각 특성의 이름)

plot_tree(DT, max_depth = 2, filled = True, feature_names=['x', 'y'])
```

#아이리시

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create a decision tree classifier
clf = DecisionTreeClassifier(max_depth = 2)
clf.fit(X, y)

# Plot the decision tree
fig, ax = plt.subplots(figsize=(12, 12))
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```
