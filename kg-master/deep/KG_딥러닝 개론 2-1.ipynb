#미션 lris 품종 데이터로 분류 모델 만들기

```python
import pandas as pd
import tensorflow as tf
```

```python
ir = pd.read_csv('/content/iris.csv')
ir = pd.get_dummies(ir)
print(ir)
print(ir.columns)
```

```python
invar = ir[['꽃잎길이','꽃잎폭','꽃받침길이','꽃받침폭']]
tarvar = ir[['품종_setosa','품종_virginica','품종_versicolor']]
print(ir.shape)
```

```python
x = tf.keras.layers.Input(shape=4)
y = tf.keras.layers.Dense(3, activation='softmax')(x)
model = tf.keras.models.Model(x, y)
model.compile(loss='categorical_crossentropy', metrics ='accuracy') # accuracy : 정확도
model.fit(invar, tarvar, epochs = 1000, verbose=0)
```

```python
model.predict(invar[0:5])
```

```python
tarvar[0:5]
```

```python
model.get_weights()
```

# 2번째 데이터

```python
bo = pd.read_csv('/content/boston.csv')
bo = pd.get_dummies(bo)

print(bo.columns)
print(bo.shape)
```

```python
in_vo = bo[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
ou_vo = bo[['medv']]
```

```python
x = tf.keras.layers.Input(shape=13)
y = tf.keras.layers.Dense(1, activation='softmax')(x)
model = tf.keras.models.Model(x,y)
model.compile(loss='categorical_crossentropy', metrics ='accuracy')
model.fit(in_vo, ou_vo, epochs=1000, verbose=0)
```

```python
model.fit(in_vo, ou_vo, epochs=1000)
```

```python
model.predict(in_vo[0:5])
```

```python
model.get_weights()
```
