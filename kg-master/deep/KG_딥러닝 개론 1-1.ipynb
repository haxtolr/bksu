#matplotlib 의 기초


```python
import matplotlib, matplotlib.pyplot as plt
```

```python
x = df['number']
y = df['height']
```

```python
plt.plot(x, y) # x - y 그래프
plt.show() # 그래프 출력
```

```python
plt.plot(x,y)
plt.axis([0, 5, 170, 180]) # x: 0~5, y:170 ~ 180 범위 지정
plt.show()
```

```python
plt.plot(x,y)
plt.axis([0, 20, 100, 200])
```

```python
plt.plot(x,y, 'k') # k:black, g : green, r : red
```

```python
plt.plot(x, y, 'r', x, y+10, 'o') # 'o'는 점 표시 산점도
```

```python
plt.plot(x, y, 'r*', x, y+10, 'go') # r* : 빨간별 산점도, go : 초록점 산점
```

```python
plt.plot(x, y, 'r', label='height1') #범례(legend)
plt.plot(x, y+10, 's', label='height2')
plt.legend()
plt.xlabel('number')
plt.ylabel('a')
plt.show()
```

```python
plt.plot(x, y, 'g', label='height1') #범례(legend)
plt.plot(x, y+10, 's', label='height2')
plt.legend(loc='best', fontsize=20) #loc:location
plt.xlabel('number')
plt.ylabel('height')
plt.show()
```

```python
plt.scatter(x,y, c = '#E389E9')
```

```python
plt.scatter(x,y, c = '#E389E9', label = 'height1')
plt.scatter(x, y+10, label= 'height2')
plt.legend(loc='best', fontsize=15)
plt.xlabel('number')
plt.ylabel('height')
```
